```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load

raw_file_path = Path("Resources/Conditions_Contributing_to_COVID-19_Deaths__by_State_and_Age__Provisional_2020-2023.csv")

# Read the raw data file and store into pandas dataframe

raw_covid_data = pd.read_csv(raw_file_path)
raw_covid_data.head(5)
```

```python
# Show the overview of the columns

raw_covid_data.columns
```

```python
# Creata a dataframe with the necessary columns to be used for data analysis

filtered_covid_data = raw_covid_data[['Year', 'Month', 'Group', 'State', 'Condition', 'Age Group', 'COVID-19 Deaths']]
filtered_covid_data
```

```python
# Export the filtered covid data to be used for analysis

filtered_covid_data.to_csv("Resources/filtered_covid_data.csv", index = False)
```

```python
# Filter the csv based on these parameters:
# Condition: COVID-19
# Age Group: Filter out "All Ages" and "Not stated"
# Group: Show data "By Month"
# States: Filter out "New York City", "District of Columbia", and "Puerto Rico"

further_filtered_covid_data = filtered_covid_data[
    (filtered_covid_data["Condition"] == "COVID-19") &
    (filtered_covid_data["Group"] == "By Month") &
    ~(filtered_covid_data["State"].isin(["District of Columbia", "New York City", "Puerto Rico", "United States"])) &
    ~(filtered_covid_data["Age Group"].isin(["All Ages", "Not stated"]))
]

further_filtered_covid_data
```

```python
# Export the further filtered datafram that has the mentioned parameters

further_filtered_covid_data.to_csv("Resources/further_filtered_covid_data.csv", index = False)
```

```python
# Now that the data is showing only COVID-19 related deaths, we can delete that column as it is repeating

further_filtered_wo_condition = further_filtered_covid_data.drop("Condition", axis = 1)
further_filtered_wo_condition
```

```python
# dropna the values in covid-19 deaths column that do not show a value

further_filtered_wo_condition = further_filtered_wo_condition.dropna(subset = ['COVID-19 Deaths'])
further_filtered_wo_condition
```

```python
# Change the datatypes of the dataframe without the condition column

# Change the float64 datatypes to integers

further_filtered_wo_condition['Year'] = further_filtered_wo_condition['Year'].astype(int)
further_filtered_wo_condition['Month'] = further_filtered_wo_condition['Month'].astype(int)
further_filtered_wo_condition['COVID-19 Deaths'] = further_filtered_wo_condition['COVID-19 Deaths'].astype(int)

print(further_filtered_wo_condition.dtypes)
```

```python
further_filtered_wo_condition
```

```python
# Export this new dataframe without the conditons column and the blank values within COVID-19 Deaths

further_filtered_wo_condition.to_csv("Resources/further_filtered_wo_condition.csv", index = False)
```

```python
# Filter the csv based on these parameters:
# Condition: COVID-19
# Age Group: Show data for "All Ages"
# Group: Show data "By Year"
# States: Filter out "New York City", "District of Columbia", "Puerto Rico", "United States"

covid_data_year = filtered_covid_data[
    (filtered_covid_data["Condition"] == "COVID-19") &
    (filtered_covid_data["Group"] == "By Year") &
    (filtered_covid_data["Age Group"] == "All Ages") &
    ~(filtered_covid_data["State"].isin(["District of Columbia", "New York City", "Puerto Rico", "United States"]))
]

covid_data_year
```

```python
# Export the year dataframe

covid_data_year.to_json("Resources/covid_data_year.json", index = False)
```

```python
# Create dataframe to use for heatmap
# The heatmap will show the top and bottom states based on their covid deaths
```

```python
# Make a dataframe showing covid deaths in each state per year

covid_data_per_year = filtered_covid_data[
    (filtered_covid_data["Condition"] == "COVID-19") &
    (filtered_covid_data["Group"] == "By Year") &
    (filtered_covid_data["Age Group"] == "All Ages") &
    ~(filtered_covid_data["State"].isin(["District of Columbia", "New York City", "Puerto Rico", "United States"]))
]

covid_data_per_year   
```

```python
# Delete unnecessary columns after the filtering

covid_data_per_year_df = pd.DataFrame(covid_data_per_year)

covid_data_per_year_df.drop(columns = ['Month', 'Group', 'Age Group', 'Condition'], inplace = True)

covid_data_per_year_df
```

```python
# Export dataframe with covid deaths in each state per year

#covid_data_per_year_df.to_csv("Resources/covid_data_per_year_df.csv", index = False)
covid_data_per_year_df.to_json("Resources/covid_data_per_year_df.json", index = False)
```
